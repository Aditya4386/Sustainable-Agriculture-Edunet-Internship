```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
dataset['Soil_color'] = pd.DataFrame(le.fit_transform(dataset['Soil_color']))
dataset['Crop'] = pd.DataFrame(le.fit_transform(dataset['Crop']))
dataset['District_Name'] = pd.DataFrame(le.fit_transform(dataset['District_Name']))
```

```python
dataset.head()
```

```python
dataset['Fertilizer'] = pd.Series(le.fit_transform(dataset['Fertilizer']))
dataset.head()
```

# Crop Prediction Model

```python
x = dataset.iloc[:, :-3]
y = dataset.iloc[:, -3]
```

```python
x.head(5)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=20)
```

```python
dt.fit(x_train, y_train)
```

```python
print(dt.score(x_train, y_train) * 100)
print(dt.score(x_test, y_test) * 100)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = dt.predict(x_test)

print(classification_report(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='g')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import joblib

# 1. Load the original dataset
original_dataset = pd.read_csv("Crop and fertilizer dataset.csv")

# 2. Create a copy for label encoding
encoded_dataset = original_dataset.copy()

# 3. Apply Label Encoding to the copy
le_district = LabelEncoder().fit(original_dataset['District_Name'])
encoded_dataset['District_Name'] = le_district.transform(original_dataset['District_Name'])

le_soil = LabelEncoder().fit(original_dataset['Soil_color'])
encoded_dataset['Soil_color'] = le_soil.transform(original_dataset['Soil_color'])

le_crop = LabelEncoder().fit(original_dataset['Crop'])
encoded_dataset['Crop'] = le_crop.transform(original_dataset['Crop'])

le_fertilizer = LabelEncoder().fit(original_dataset['Fertilizer'])
encoded_dataset['Fertilizer'] = le_fertilizer.transform(original_dataset['Fertilizer'])

# --- MODEL 1: CROP PREDICTION ---
print("Training Crop Prediction Model...")
x_crop = encoded_dataset.iloc[:, :-3]  # Features: All columns except Crop, Fertilizer, Link
y_crop = encoded_dataset.iloc[:, -3]   # Target: The 'Crop' column
x_crop_train, x_crop_test, y_crop_train, y_crop_test = train_test_split(x_crop, y_crop, train_size=0.8, random_state=42)

# Create and train the model
dt_crop = DecisionTreeClassifier(criterion='entropy', max_depth=20)
dt_crop.fit(x_crop_train, y_crop_train)
print("Crop Model Trained Successfully!")


# --- MODEL 2: FERTILIZER PREDICTION ---
print("\\nTraining Fertilizer Prediction Model...")
x_fertilizer = encoded_dataset.iloc[:, :-2] # Features: All columns except Fertilizer, Link
y_fertilizer = encoded_dataset.iloc[:, -2]  # Target: The 'Fertilizer' column
x_fert_train, x_fert_test, y_fert_train, y_fert_test = train_test_split(x_fertilizer, y_fertilizer, train_size=0.8, random_state=42)

# Create and train the model
dt_fertilizer = DecisionTreeClassifier(criterion='entropy', max_depth=20)
dt_fertilizer.fit(x_fert_train, y_fert_train)
print("Fertilizer Model Trained Successfully!")


# --- SAVE EVERYTHING ---
print("\\nSaving all models, encoders, and mappings...")

# Save the two models
joblib.dump(dt_crop, 'crop_prediction_model.pkl')
joblib.dump(dt_fertilizer, 'fertilizer_prediction_model.pkl')

# Save all encoders
joblib.dump(le_district, 'district_name_encoder.pkl')
joblib.dump(le_soil, 'soil_color_encoder.pkl')
joblib.dump(le_crop, 'crop_encoder.pkl')
joblib.dump(le_fertilizer, 'fertilizer_encoder.pkl')

# Save the YouTube link mapping
fertilizers_df = original_dataset[['Crop', 'Link']].drop_duplicates().reset_index(drop=True)
mapping = fertilizers_df.set_index('Crop')['Link'].to_dict()
joblib.dump(mapping, 'crop_to_link_mapping.pkl')

print("\\n✅ All files saved successfully!")
```
